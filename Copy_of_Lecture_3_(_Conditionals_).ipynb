```python
#WAP to Find Gratest Among 3 numbers

n1= int(input("Enter First No.: "))
n2=int(input("ENter Second No.: "))
n3=int(input("ENter Third No.: "))

if(n1>n2 & n1>n3):
  print(n1, "Is Gratest")
elif(n2>n1 & n2>n3):
  print(n2, "Is Gratest")
else:
  print(n3, "Is Gratest")

```
