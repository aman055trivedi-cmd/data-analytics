```python
3#WAP to print Area of triangle otherwise print not equilateral triangle
import math
# Input sides of triangle
a =float(input("Enter the length of side a: "))
b =float(input("Enter the length of side b: "))
c =float(input("Enter the length of side c: "))

# Check if triangle is equilateral
if a == b == c:
    area = (math.sqrt(3) / 4) * a * a
    print(f"Area of equilateral triangle: {area:.2f}")
else:
    print("Not an equilateral triangle")
```

```python
#wap  using membership operater and identity operate
# Membership Operator Example
fruits = ['apple', 'banana', 'cherry']
fruit_to_check = 'banana'

if fruit_to_check in fruits:
    print(f"{fruit_to_check} is in the list of fruits.")
else:
    print(f"{fruit_to_check} is not in the list of fruits.")

# Identity Operator Example
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print("x is y:", x is y)
print("x is z:", x is z)
print("x == z:", x == z)
```
