Coin Toss Game



1. Toss a coin 100000 times and find out the probability of being head and tail

```python
import random

tosses = 100000
heads = 0

tails = 0

for _ in range(tosses):
    if random.choice(['H', 'T']) == 'H':
        heads += 1
    else:
        tails += 1

print("Heads:", heads)
print("Tails:", tails)
print("Probability of Head:", heads / tosses)
print("Probability of Tail:", tails / tosses)
```

```python
class Rectangle:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  def calculate_area(self):
    return self.length * self.breadth

  def calculate_perimeter(self):
    return 2 * (self.length + self.breadth)


rect1 = Rectangle(10, 5)

print(f"Rectangle 1 - Length: {rect1.length}, Breadth: {rect1.breadth}")
print(f"Area: {rect1.calculate_area()}")
print(f"Perimeter: {rect1.calculate_perimeter()}")

rect2 = Rectangle(7.5, 3.2)
print(f"\nRectangle 2 - Length: {rect2.length}, Breadth: {rect2.breadth}")
print(f"Area: {rect2.calculate_area()}")
print(f"Perimeter: {rect2.calculate_perimeter()}")
```

```python
import numpy as np

# Create an empty 5x5 matrix
matrix = np.zeros((5, 5), dtype=int)

# Populate the first column with increasing order
matrix[:, 0] = np.arange(1, 6) # Values from 1 to 5

# Populate the last column with decreasing order
matrix[:, -1] = np.arange(5, 0, -1) # Values from 5 down to 1

# Fill the middle columns with random integers (e.g., between 1 and 10) for illustration
matrix[:, 1:-1] = np.random.randint(1, 11, size=(5, 3)) # 5 rows, 3 middle columns

print("Generated 5x5 Matrix:")
print(matrix)
```

```python
import numpy as np

# Assuming 'matrix' from the previous cell is available in the environment
# If not, you might need to re-run the previous cell or define matrix here.

# Calculate the mean of the entire matrix
mean_matrix = np.mean(matrix)

# Calculate the standard deviation of the entire matrix
std_dev_matrix = np.std(matrix)

# Apply the Z-score formula: (value - mean) / standard_deviation
# Avoid division by zero if standard deviation is 0 (e.g., if all values are the same)
if std_dev_matrix == 0:
    z_score_matrix = np.zeros_like(matrix)
else:
    z_score_matrix = (matrix - mean_matrix) / std_dev_matrix

print("Original 5x5 Matrix:")
print(matrix)
print("\nZ-score transformed 5x5 Matrix:")
print(z_score_matrix)
```

```python
import numpy as np
from scipy.stats import zscore

# Let's use the 'matrix' we created earlier
# If 'matrix' is not defined, you can create a sample array:
# matrix = np.array([[1, 7, 8, 6, 5],
#                    [2, 2, 7, 9, 4],
#                    [3, 6, 5, 2, 3],
#                    [4, 8, 9, 7, 2],
#                    [5, 1, 4, 4, 1]])

print("Original Matrix:")
print(matrix)

# Calculate Z-scores for the entire matrix
z_score_transformed_matrix = zscore(matrix)

print("\nZ-score transformed Matrix using scipy.stats.zscore:")
print(z_score_transformed_matrix)

# You can also calculate Z-scores along a specific axis (e.g., column-wise or row-wise)
# Z-score column-wise (axis=0)
z_score_column_wise = zscore(matrix, axis=0)
print("\nZ-score transformed (column-wise) Matrix:")
print(z_score_column_wise)

# Z-score row-wise (axis=1)
z_score_row_wise = zscore(matrix, axis=1)
print("\nZ-score transformed (row-wise) Matrix:")
print(z_score_row_wise)
```

```python
import numpy as np

actual_array = np.array([10, 12, 15, 11, 13, 14, 17, 9, 7, 2])
predicted_array = np.array([11, 11, 14, 12, 12, 5, 3, 4, 16, 1])
element_wise_difference = actual_array - predicted_array
mse = np.mean((actual_array - predicted_array)**2)
mae = np.mean(np.abs(actual_array - predicted_array))
print("Actual array:", actual_array)
print("Predicted array:", predicted_array)
print("Element-wise difference (actual - predicted):", element_wise_difference)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

```python
import numpy as np

# Define the shape for the 3D array
shape = (2, 3, 4)

print("--- Method 1: Using np.full to fill with a specific boolean value ---")
# np.full(shape, fill_value, dtype=bool)
array_full_true = np.full(shape, True, dtype=bool)
print("Array filled with True:")
print(array_full_true)

print("\n--- Method 2: Using np.zeros and specifying dtype=bool ---")
# np.zeros(shape, dtype=bool) creates an array of False values
array_zeros_bool = np.zeros(shape, dtype=bool)
print("Array filled with False:")
print(array_zeros_bool)

print("\n--- Method 3: Boolean comparison on a numerical array (e.g., from np.arange) ---")
# Create a numerical array and then apply a boolean condition
numerical_array_arange = np.arange(np.prod(shape)).reshape(shape)
array_from_comparison_arange = numerical_array_arange % 2 == 0 # True for even numbers
print("Array from numerical comparison (even numbers are True):")
print(array_from_comparison_arange)

print("\n--- Method 4: Boolean comparison on a random numerical array (e.g., from np.random.randint) ---")
# Create a random numerical array and then apply a boolean condition
np.random.seed(42) # for reproducibility
numerical_array_random = np.random.randint(0, 10, size=shape)
array_from_comparison_random = numerical_array_random > 5 # True for values greater than 5
print("Array from random numerical comparison (values > 5 are True):")
print(array_from_comparison_random)
```
