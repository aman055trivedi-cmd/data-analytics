```python
#WAP TO IMPLEMENT LINIER SEARCH IN PHYTHON
def linear_search(arr, target):
    """
    Searches for a target value in an array using linear search.

    Args:
        arr: A list of elements.
        target: The element to search for.

    Returns:
        The index of the target if found, otherwise -1.
    """
    # Loop through each element and its index in the array
    for i in range(len(arr)):
        # Check if the current element matches the target
        if arr[i] == target:
            return i  # Return the index if a match is found

    return -1  # Return -1 if the loop finishes without finding the target

# --- Example Usage ---
my_list = [10, 50, 30, 70, 80, 60, 20]
search_item = 80

result = linear_search(my_list, search_item)

if result != -1:
    print(f"Element {search_item} found at index: {result}")
else:
    print(f"Element {search_item} not found in the list.")

# --- Output ---
# Element 80 found at index: 4
```

```python
# 1. String Index Error: Write a function char_at(s, i) that returns the character at position i in string s Handle IndexError and TypeError
def char_at(s, i):
    try:
        return s[i]
    except IndexError:
        return "IndexError: Index out of range"
    except TypeError:
        return "TypeError: Invalid input type"
```

```python
#2. String Conversion: Write code that converts a string input to integer. Handle the case when the string is not numeric using ValueError.
def convert_to_int(s):
    try:
        return int(s)
    except ValueError:
        return "ValueError: Input string is not a valid integer"

# Example usage:
print(convert_to_int("123"))
print(convert_to_int("abc123"))
```

```python
# 3. Divide List Elements: Define a function divide_elements(L, n) to divide each element of list L by number n. Handle ZeroDivisionError and TypeError.
def divide_elements(L, n):
    try:
        return [x / n for x in L]
    except ZeroDivisionError:
        return "ZeroDivisionError: Cannot divide by zero"
    except TypeError:
        return "TypeError: Invalid input type"

# Example usage:
print(divide_elements([60, 70, 80], 55))
print(divide_elements([60, 70, 80], 0))
print(divide_elements("not a list", 55))
```

```python
# 7. Dictionary Key Lookup: Create a function find_value(d, key) that returns the value for a given key
class KeyNotFoundError(Exception):
    def __init__(self, key):
        super().__init__(f"KeyNotFoundError: '{key}' not found in the dictionary.")

# Function to update dictionary
def update_if_exists(d, key, value):
    if key in d:
        d[key] = value
        print(f"Updated '{key}' to '{value}'")
    else:
        raise KeyNotFoundError(key)

# Example usage
my_dict = {"name": "Aman", "age": 25, "city": "Kanpur"}

try:
    update_if_exists(my_dict, "age", 26)
    update_if_exists(my_dict, "country", "India")
except KeyNotFoundError as e:
    print(e)
```

```python
#-8 Dictionary Update: Write a function that updates a dictionary only if the key already exists, otherwise raises a custom exception KeyNotFoundError
class KeyNotFoundError(Exception):
    def __init__(self, key):
        super().__init__(f"KeyNotFoundError: '{key}' not found in the dictionary.")

# Function to update dictionary
def update_if_key_exists(d, key, value):
    if key in d:
        d[key] = value
        print(f"✅ Updated '{key}' to '{value}'")
    else:
        raise KeyNotFoundError(key)

# Example usage
student_info = {"name": "Aman", "roll_no": 1023, "year": 2025}

try:
    update_if_key_exists(student_info, "year", 2026)      # ✅ Key exists
    update_if_key_exists(student_info, "gender", "Male")  # ❌ Key doesn't exist
except KeyNotFoundError as e:
    print(e)

```

```python
#13
def divide_list_elements(data):
    try:
        for item in data:
            try:
                print(item / 2)
            except TypeError:
                print(f"TypeError: Cannot divide {item} by 2.")
    except TypeError:
        print("Outer TypeError: Provided input is not a list.")
```

```python
#15
def read_file_safely(filename):
    file = None
    try:
        file = open(filename, "r")
        print(file.read())
    except FileNotFoundError:
        print("FileNotFoundError: File not found.")
    finally:
        if file:
            file.close()
            print("File closed.")
```

```python
#16
def check_age(age):
    try:
        if age < 18:
            raise ValueError("Age must be 18 or older.")
        print("Access granted.")
    except ValueError as e:
        print(f"ValueError: {e}")
```

```python
#17. Define custom exception
class NegativeNumberError(Exception):
    def __init__(self, number):
        super().__init__(f"NegativeNumberError: {number} is negative.")

# Function to check positivity
def check_positive(n):
    if n < 0:
        raise NegativeNumberError(n)
    print(f"{n} is positive.")

# Example usage
try:
    check_positive(5)
    check_positive(-3)
except NegativeNumberError as e:
    print(e)
```

```python
#19.
def set_intersection(a, b):
    try:
        return a.intersection(b)
    except AttributeError:
        print("AttributeError: Both inputs must be sets.")

# Example usage
print(set_intersection({1, 2, 3}, {2, 3, 4}))
print(set_intersection([1, 2], [2, 3]))

```

```python
#20.
def square(x):
    return x * x

def main():
    try:
        num = input("Enter a number: ")
        result = square(float(num))
        print("Square:", result)
    except ValueError:
        print("ValueError: Please enter a numeric value.")
    except TypeError:
        print("TypeError: Invalid input type.")

# Run the main function
main()

```
